```
import heapq
import time

# Define the goal state
goal_state = [[1, 2, 3], [4, 5, 6], [7, 8, 0]]

# Define the Manhattan distance heuristic
def mahatten_distance(state):
    distance=0
    for i in range(3):
        for j in range(3):
            if state[i][j] !=0:
                for x in range(3):
                    for y in range(3):
                        if state[i][j]==goal_state[x][y]:
                            distance+=abs(i-x)+abs(y-j)
    return distance

# Define the misplaced tiles heuristic
def misplaced_tiles(state):
    count = 0
    for i in range(3):
        for j in range(3):
            if state[i][j] !=0 and state[i][j] != goal_state[i][j]:
                count += 1
    return count

# Define the A* algorithm without heuristic
def astar_no_heuristic(initial_state):
    return 0

# Define the non-optimal and inadmissible heuristic
def constant_heuristic(state):
    return 10

# Helper functions
def find_zero_position(state):
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                return i, j
# function to swap val of state
def swap(state, row1, col1, row2, col2):
    new_state = [row[:] for row in state]
    new_state[row1][col1], new_state[row2][col2] = new_state[row2][col2], new_state[row1][col1]
    return new_state

# Define the A* algorithm with Manhattan distance heuristic
def astar(initial_state, heuristic):
    open_list = []
    closed_list = set()
    explored_states = 0

    # Create the initial node
    initial_g = 0
    initial_h = heuristic(initial_state)
    # initial_h = 0

    heapq.heappush(open_list, (initial_g + initial_h, initial_g, initial_state, []))

    # Start exploring nodes
    while open_list:
        _, g, current_state, path = heapq.heappop(open_list)
        #print(current_state,_)
        closed_list.add(tuple(map(tuple, current_state)))

        if current_state == goal_state:
            return path, len(closed_list)

        # Generate child nodes and cal new value f = g(real cost) + h(heuristic cost)
        zero_row, zero_col = find_zero_position(current_state)
        for dr, dc in [(0, -1), (0, 1), (-1, 0), (1, 0)]:
            new_row, new_col = zero_row + dr, zero_col + dc
            if 0 <= new_row < 3 and 0 <= new_col < 3:
                #swapping val of current state
                new_state = swap(current_state, zero_row, zero_col, new_row, new_col)
                new_state_tuple = tuple(map(tuple, new_state))
                #cheking state are explored or not
                if new_state_tuple not in closed_list:
                    new_g = g + 1
                    new_h = heuristic(new_state) #calling different heuristic function
                    new_path = path + [(current_state, new_state)]
                    heapq.heappush(open_list, (new_g + new_h, new_g, new_state, new_path))

    # If no solution is found
    return None, len(closed_list)

# Test the algorithm with Manhattan distance heuristic

initial_state = [[3, 2, 1], [4, 5, 6], [8, 7, 0]]




if __name__ == "__main__":
  """
  calling diffrent type of heuristic function
  """
  for heuristic in [astar_no_heuristic, misplaced_tiles, mahatten_distance, constant_heuristic]:
    start_time = time.time() #start time
    path, num_states_explored = astar(initial_state, heuristic)
    end_time = time.time() #end time
    print(f'********* START *************')
    print(f' ${heuristic} ')
    if path:
        print("Goal node is reachable!")
        print(f"Time taken: {end_time - start_time} seconds") #time taken by algo
        print(f"Total states explored: {num_states_explored}")
        print("Initial state:")
        print(initial_state)
        print("Goal state:")
        print(goal_state)
    else:
        print("Goal node is unreachable!")
    print(f'********* END *************')
```
